```python
from pyhht.emd import EmpiricalModeDecomposition as EMD
from pyhht.visualization import plot_imfs
import numpy as np
```

```python
x = np.arange(-10, 10, 0.1)
f1 = np.sin(1/x) - np.cos(x**2)
y1 = EMD(f1).decompose()
plot_imfs(f1, y1)
```

```python
f2 = np.tanh(x) + np.cos(x**2)
y2 = EMD(f2).decompose()
plot_imfs(f2, y2)
```
